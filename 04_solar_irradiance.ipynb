```python
import requests
import pandas as pd
import numpy as np
from datetime import datetime

# Select a central point in Accra 
lat = 5.6037
lon = -0.1869

# NASA POWER API - Daily data for the last 5+ years
url = (
    f"https://power.larc.nasa.gov/api/temporal/daily/point?"
    f"parameters=ALLSKY_SFC_SW_DWN&"
    f"community=RE&"
    f"longitude={lon}&"
    f"latitude={lat}&"
    f"start=20200101&"
    f"end=20251231&"
    f"format=CSV"
)

print("Fetching solar irradiance data from NASA POWER...")

response = requests.get(url)
if response.status_code == 200:
    # Skip header lines (NASA CSV has 2 header rows)
    lines = response.text.splitlines()
    data_start = next(i for i, line in enumerate(lines) if line.startswith("YEAR"))
    
    df = pd.read_csv(url, skiprows=data_start)
    
    print(f"Data fetched successfully for period: {df['YEAR'].min()}-{df['YEAR'].max()}")
    
    # ALLSKY_SFC_SW_DWN = Global Horizontal Irradiance (kWh/m²/day)
    ghi_daily = df['ALLSKY_SFC_SW_DWN']
    
    annual_ghi = ghi_daily.mean() * 365.25   # Approximate annual sum
    print(f"\nAverage Daily GHI: {ghi_daily.mean():.2f} kWh/m²/day")
    print(f"Estimated Annual GHI: {annual_ghi:.0f} kWh/m²/year")
    
    # Save the data
    df.to_csv("accra_nasa_power_ghi.csv", index=False)
    print("Data saved as accra_nasa_power_ghi.csv")
    
else:
    print(f"Error fetching data: {response.status_code}")
    print(response.text)
```
