```python
# Load Combined Dataset and Filter to Accra

import ee
import geopandas as gpd
from shapely.geometry import box

# Define central Accra bounding box
accra_box = box(-0.25, 5.50, 0.00, 5.65)

# Load VIDA Combined (Google + Microsoft) for Ghana only
buildings = ee.FeatureCollection("projects/sat-io/open-datasets/VIDA_COMBINED/GHA")

# Filter to our pilot area
accra_buildings_ee = buildings.filterBounds(
    ee.Geometry.Rectangle([-0.25, 5.50, 0.00, 5.65])
)

print(f"Number of buildings in filter (preview): {accra_buildings_ee.size().getInfo()}")

print("\nExporting buildings from GEE to Google Drive... (this may take 30 seconds to 3 minutes)")

# Export as GeoJSON
task = ee.batch.Export.table.toDrive(
    collection=accra_buildings_ee,
    description='accra_combined_buildings_GHA',
    fileFormat='GeoJSON',
    folder='EarthEngineExports'
)

task.start()

print("Export task started successfully!")
```

```python
gdf_combined = gpd.read_file("accra_combined_buildings_GHA.geojson")

print(f"Total buildings loaded: {len(gdf_combined):,}")
if 'bf_source' in gdf_combined.columns:
    print("Source breakdown (Google vs Microsoft):")
    print(gdf_combined['bf_source'].value_counts(dropna=False))

# Save
gdf_combined.to_file("accra_buildings_combined.gpkg", driver="GPKG")
print("Saved as accra_buildings_combined.gpkg")
```
